```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
driver.get('https://www.essexapartmenthomes.com/')
```

```python
driver.find_element(By.XPATH,'//*[@class="text-card__columns__col"]/span/a').click()
```

```python
button = driver.find_element(By.XPATH, '//a[contains(@href, "Northern California")]/span[text()="Northern California"]')
button.click()
time.sleep(5)
```

```python
try:
    see_details_elements = driver.find_elements(By.XPATH, '//a[@class="button-primary ga-nav"]') 
    see_details_urls = [element.get_attribute("href") for element in see_details_elements]

    for url in see_details_urls:
        print(url)
        
except Exception as e:
    print(f"Failed to extract 'See Details' links: {e}")
print(len(see_details_urls))
```

```python
records = []

driver.maximize_window()

for url in see_details_urls:
    driver.get(url)
    time.sleep(3)
    
    record = {}

    time.sleep(3)
    community_text = driver.find_element(By.XPATH, '//*[@class="community-name-block"]/h1').text
    record['community_title'] = community_text
    
    time.sleep(3)
    price_ele = driver.find_elements(By.XPATH, '//*[@class="price-container"]/span[@class="price"]')   
    prices = [first_text.text for first_text in price_ele]
    record['price'] = prices
    
    time.sleep(3)
    Address_text = driver.find_element(By.XPATH, '//*[@rel="noopener noreferrer"]/p').text
    record['city'] = Address_text
    
    driver.find_element(By.XPATH, '//*[@class="navigation__category"][2]/a').click()
    time.sleep(3)
    
    floor_elements = driver.find_elements(By.XPATH, '//*[@class="floor-plan-card__content__size"]')
    floors = [floor_element.text for floor_element in floor_elements]
    record['floor_details'] = floors
    
    time.sleep(3)
    price_elements = driver.find_elements(By.XPATH, '//*[@class="floor-plan-card__content__price"]')
    prices = [price_element.text for price_element in price_elements]
    record['price_details'] = prices
    
    time.sleep(3)
    driver.back()
    time.sleep(3)
    # Getting neighborhood details
    driver.find_element(By.XPATH, '//*[@class="navigation__category"][3]/a').click()
    time.sleep(3)
    
    score_text = driver.find_element(By.XPATH, '//*[@class="score-content-left-desc walk-score"]').text
    record['score'] = score_text

    driver.back()
    time.sleep(3)
    
    records.append(record)
```
